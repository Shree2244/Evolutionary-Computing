## **Heuristic methods (First Fit, Best Fit)**

```python
def first_fit(tasks, vms):
    task_schedule = []
    for task in tasks:
        for vm in vms:
            if vm.cpu.capacity >= task.cpu and vm.ram >= task.memory:
                task_schedule.append((task, vm))
                vms.remove(vm)  # Mark VM as unavailable
                break
        else:
            print("Warning: No available VM to accommodate the task:", task.name)
    return task_schedule

def calculate_completion_time(task_schedule):
    completion_time = 0
    for task, vm in task_schedule:
        completion_time += task.time / min(vm.cpu.capacity / task.cpu, vm.ram / task.memory)
    return completion_time

# Execute the heuristic algorithm
task_schedule = first_fit(tasks, vms)

def warn_no_vm_available(tasks, vms):
    for task in tasks:
        if all(not (vm.cpu.capacity >= task.cpu and vm.ram >= task.memory) for vm in vms):
            print("Warning: No available VM to accommodate the task:", task.name)


# Call the warning function
warn_no_vm_available(tasks, vms)

# Calculate completion time and print results
hm_completion_time = calculate_completion_time(task_schedule)
print("Task Schedule:")
for task, vm in task_schedule:
    print(f"Task: {task.name}, VM: {vm.name}")
print("Total Completion Time:", hm_completion_time)
```

```python
def best_fit(tasks, vms):
    task_schedule = []
    for task in tasks:
        best_vm = None
        best_score = float('inf')
        for vm in vms:
            if vm.cpu.capacity >= task.cpu and vm.ram >= task.memory:
                score = max(vm.cpu.capacity / task.cpu, vm.ram / task.memory)
                if score < best_score:
                    best_score = score
                    best_vm = vm
        if best_vm is not None:
            task_schedule.append((task, best_vm))
            vms.remove(best_vm)
        else:
            print("Warning: No available VM to accommodate the task:", task.name)
    return task_schedule

def calculate_completion_time(task_schedule):
    completion_time = 0
    for task, vm in task_schedule:
        completion_time += task.time / min(vm.cpu.capacity / task.cpu, vm.ram / task.memory)
    return completion_time


# Calculate completion time and print results
bf_completion_time = calculate_completion_time(task_schedule)

print("\nBest Fit Task Schedule:")
for task, vm in task_schedule:
    print(f"Task: {task.name}, VM: {vm.name}")
print("Best Fit Total Completion Time:", bf_completion_time)
```

## **Genetic Algorithm**

```python
#parameters
population_size = 10
max_generations = 20
mutation_rate = 0.1
```

```python
def generate_chromosome(vms, tasks):
    return [random.choice(vms) for _ in range(len(tasks))] # generating random task schedule

def fitness(chromosome):
    completion_time = 0 #fitness function is total completion time
    for i, task in enumerate(tasks):
        vm = chromosome[i]
        completion_time += task.time / min(vm.mips / task.cpu, vm.ram / task.memory)# Calculating total completion time for each task schedule
    return completion_time

def crossover(parent1, parent2, tasks): # single-point crossover
    crossover_point = random.randint(1, len(tasks) - 1)
    child1 = parent1[:crossover_point] + parent2[crossover_point:]
    child2 = parent2[:crossover_point] + parent1[crossover_point:]
    return child1, child2


def mutate(chromosome, tasks, mutation_rate): #uniform mutation (mutation_rate = 0.1)
    mutated_chromosome = chromosome.copy()
    for i in range(len(chromosome)):
        if random.random() < mutation_rate:
            mutated_chromosome[i] = random.choice(vms)
    return mutated_chromosome

def genetic_algorithm(vms, tasks):
    population = [generate_chromosome(vms, tasks) for _ in range(population_size)] # Initializing the population

    for generation in range(max_generations):
        fitness_values = [fitness(chromosome) for chromosome in population] # Calculating fitness for each chromosome

        # Parents selection (for crossover)
        parents = random.choices(population, weights=[1/fit for fit in fitness_values], k=2)

        # Performing crossover and mutation
        offspring = []
        for _ in range(population_size // 2):
            child1, child2 = crossover(parents[0], parents[1], tasks)
            offspring.extend([mutate(child1, tasks, mutation_rate), mutate(child2, tasks, mutation_rate)])

        # Survivor selection
        population = offspring

        # Best solutions with highest fitness, minimizing
        best_solution = min(population, key=fitness)
        best_vm_names = [vm.name for vm in best_solution]  # Extract VM names
        print(f"Generation {generation + 1}, Best Solution: {best_vm_names}, Fitness: {fitness(best_solution)}")

    return min(population, key=fitness)
```

```python
def calculate_resource_utilization(solution):
    total_cpu_capacity = sum(vm.mips for vm in solution) # Calculating resource utilization - cpu
    total_memory_capacity = sum(vm.ram for vm in solution) # Calculating resource utilization - memory
    used_cpu = sum(task.cpu for task in tasks)
    used_memory = sum(task.memory for task in tasks)
    cpu_utilization = used_cpu / total_cpu_capacity
    memory_utilization = used_memory / total_memory_capacity
    return cpu_utilization, memory_utilization
```

```python
best_ga_solution = genetic_algorithm(vms, tasks)
print("Best Solution Found:", [vm.name for vm in best_ga_solution])
print("Total Completion Time:", fitness(best_ga_solution))
print("Resource Utilization (CPU, Memory):", calculate_resource_utilization(best_ga_solution))
```

## **Ant Colony Optimization**

```python
# Parameters
NUM_ANTS = 5
NUM_ITERATIONS = 10
EVAPORATION_RATE = 0.5
ALPHA = 1
BETA = 2
Q = 1
```

```python
def calculate_probabilities(vms, task, pheromone_matrix, vm_indices):
    probabilities = []
    total_probability = 0
    for vm_index in vm_indices:
        pheromone = pheromone_matrix[vm_index]
        heuristic = 1  # Placeholder for heuristic calculation
        probability = pheromone * heuristic
        probabilities.append(probability)
        total_probability += probability
    probabilities = [p / total_probability for p in probabilities]
    return probabilities

def update_pheromone(pheromone_matrix, ant_routes):
    for ant_route in ant_routes:
        for vm_index in ant_route:
            pheromone_matrix[vm_index] += Q

def aco(env, vms, tasks):
    pheromone_matrix = np.ones(NUM_VMS)  # Initialize pheromone matrix
    best_solution = None
    best_completion_time = float('inf')
    for iteration in range(NUM_ITERATIONS):
        ant_routes = []
        for ant in range(NUM_ANTS):
            vm_indices = list(range(NUM_VMS))
            ant_route = []
            while vm_indices:
                probabilities = calculate_probabilities(vms, tasks[ant % NUM_TASKS], pheromone_matrix, vm_indices)
                selected_vm_index = np.random.choice(vm_indices, p=probabilities)
                ant_route.append(selected_vm_index)
                vm_indices.remove(selected_vm_index)
            ant_routes.append(ant_route)
            completion_time = calculate_completion_time(ant_route, vms, tasks)
            if completion_time < best_completion_time:
                best_solution = ant_route
                best_completion_time = completion_time
        update_pheromone(pheromone_matrix, ant_routes)
    return best_solution, best_completion_time
```

```python
def calculate_completion_time(vm_indices, vms, tasks):
    completion_time = 0
    for i, vm_index in enumerate(vm_indices):
        vm = vms[vm_index]
        task = tasks[i % len(tasks)]
        completion_time += task.time / min(vm.mips / task.cpu, vm.ram / task.memory)
    return completion_time

def calculate_resource_utilization(vms, tasks):
    total_cpu = sum(vm.mips for vm in vms)
    total_ram = sum(vm.ram for vm in vms)
    used_cpu = sum(task.cpu for task in tasks)
    used_ram = sum(task.memory for task in tasks)
    cpu_utilization = used_cpu / total_cpu
    ram_utilization = used_ram / total_ram
    return cpu_utilization, ram_utilization
```

```python
def main():
    best_solution, best_completion_aco_time = aco(env, vms, tasks)
    cpu_utilization, ram_utilization = calculate_resource_utilization(vms, tasks)
    print("Best Solution Found:", best_solution)
    print("Total Completion Time:", best_completion_aco_time)
    print("CPU Utilization:", cpu_utilization)
    print("RAM Utilization:", ram_utilization)

if __name__ == "__main__":
    main()
```

## **Particle Swarm Optimization**

```python
# Parameters
NUM_VMS = 5
NUM_TASKS = 10
NUM_PARTICLES = 5
NUM_ITERATIONS = 10
MAX_SPEED = 5
C1 = 0.5
C2 = 0.5
```

```python
def evaluate_fitness(particle, vms, tasks):
    # Placeholder for fitness evaluation
    return random.randint(0, 100)

def pso(env, vms, tasks):
    global_best_solution = None
    global_best_fitness = float('inf')
    particles = []
    velocities = []
    for _ in range(NUM_PARTICLES):
        particle = np.random.permutation(len(tasks))
        particles.append(particle)
        velocity = np.random.uniform(-MAX_SPEED, MAX_SPEED, len(tasks))
        velocities.append(velocity)
    print("Particles:", particles)
    print("Velocities:", velocities)
    for iteration in range(NUM_ITERATIONS):
        for i, particle in enumerate(particles):
            fitness = evaluate_fitness(particle, vms, tasks)
            if fitness < global_best_fitness:
                global_best_solution = particle
                global_best_fitness = fitness
            for j in range(len(tasks)):
                velocity_component1 = C1 * random.random() * (global_best_solution[j] - particle[j])
                velocity_component2 = C2 * random.random() * (particles[i][j] - particle[j])
                velocities[i][j] = min(MAX_SPEED, velocities[i][j] + velocity_component1 + velocity_component2)
                particles[i][j] = max(0, min(len(tasks) - 1, particle[j] + int(velocities[i][j])))
    return global_best_solution, global_best_fitness, particles  # Return particles as well

def calculate_resource_utilization(vms, tasks):
    total_cpu = sum(vm.mips for vm in vms)
    total_ram = sum(vm.ram for vm in vms)

    used_cpu = sum(task.cpu for task in tasks)
    used_ram = sum(task.memory for task in tasks)

    cpu_utilization = used_cpu / total_cpu
    ram_utilization = used_ram / total_ram

    return cpu_utilization, ram_utilization

def calculate_total_completion_time(solution, vms, tasks):
    vm_times = [0] * len(vms)
    total_cpu_usage = [0] * len(vms)  # Track total CPU usage for each VM
    total_ram_usage = [0] * len(vms)  # Track total RAM usage for each VM

    for i, task_index in enumerate(solution):
        vm_index = i % len(vms)  # Adjust index to loop through VMs
        vm_time = vm_times[vm_index]
        task = tasks[task_index]

        task_time = task.time  # Use the task's time attribute as an approximation of length
        vm_times[vm_index] = max(vm_time, task_time) + task_time

        total_cpu_usage[vm_index] += task.cpu  # Add task CPU usage to the total CPU usage of the VM
        total_ram_usage[vm_index] += task.memory  # Add task RAM usage to the total RAM usage of the VM

    # Calculate completion time based on the VMs' total CPU usage
    max_completion_time = max(vm_times)
    return max_completion_time, total_cpu_usage, total_ram_usage

# Modify main function to use the updated calculation function
def main():
    best_solution, best_fitness, particles = pso(env, vms, tasks)
    total_completion_time, _, _ = calculate_total_completion_time(best_solution, vms, tasks)  # Use only total completion time
    cpu_utilization, ram_utilization = calculate_resource_utilization(vms, tasks)  # Pass tasks to calculate resource utilization

    print("Best Solution Found:", best_solution)
    print("Total Fitness:", best_fitness)
    print("Total Completion Time:", total_completion_time)
    print("CPU Utilization:", cpu_utilization)
    print("RAM Utilization:", ram_utilization)

if __name__ == "__main__":
    main()
```
